```python
import numpy as np
import pandas as pd
```

```python
building_areas=['donghu','gaoxinkaifaqu','honggutan','jingkaiqu','jinxian','nanchangxian','qingshanhu','qingyunpu','wanli','xihu','xinjian']
columns=['avg','max','min']
```

```python
data_list=np.zeros((len(building_areas),3))
for i in range(len(building_areas)):
    df=pd.read_json(f"./data/building_area/{building_areas[i]}.json",encoding="utf-8",orient="records")
    area_price_df=df.loc[:,"price"]
    data_list[i][0]='{:.2f}'.format(area_price_df.mean())
    data_list[i][1]=area_price_df.max()
    data_list[i][2]=area_price_df.min()
    
    
area_dfs=pd.DataFrame(data=data_list,index=building_areas,columns=columns)
area_dfs
```

```python
building_cnts=np.zeros(len(building_areas),dtype=int)
building_lists=np.zeros
for i in range(len(building_areas)):
    df=pd.read_json(f"./data/building_area/{building_areas[i]}.json",encoding="utf-8",orient="records")
    data={
        "name":list(df.name),
        "price":list(df.price)
    }
    dfs=pd.DataFrame(data=data)
    display(dfs)

    #南昌各地区楼盘个数
    building_cnts[i]=int(dfs.shape[0])
print(list(building_cnts))
```

```python
area_dfs_index_cn=["东湖","高新开发区","红谷滩","经开区","进贤","南昌县","青山湖","青云谱","湾里","西湖","新建"]
avg_=list(area_dfs.loc[:,"avg"])
max_=list(area_dfs.loc[:,"max"])
min_=list(area_dfs.loc[:,"min"])
```

```python
from pyecharts.charts import Bar,Page,Line,Pie
from pyecharts import options as opts
from pyecharts.globals import ThemeType
```

```python
page=Page(layout=Page.DraggablePageLayout,interval=10)
```

```python
bar=(
    Bar(
        init_opts=opts.InitOpts(
            width="1000px",
            height="600px",
            theme=ThemeType.ESSOS #主题
        )
    
    )
    .add_xaxis(area_dfs_index_cn)
    .add_yaxis("平均房价",avg_)
    .add_yaxis("最高房价",max_)
    .add_yaxis("最低房价",min_)
    
    .reversal_axis()
    
    #系列配置项
    .set_series_opts(
        label_opts=opts.LabelOpts(position="right")
    )
    
    #全局配置项
    .set_global_opts(
        #标题配置
        title_opts=opts.TitleOpts(
            title="南昌各地房价对比条形图"
        ),
        datazoom_opts=opts.DataZoomOpts(
            is_show=True,
            range_start=0,
            range_end=60
        )
    )
    
)
```

```python
line=(
    Line()
    .add_xaxis(area_dfs_index_cn)
    .add_yaxis("平均房价",avg_)
    .add_yaxis("最高房价",max_)
    .add_yaxis("最低房价",min_)
    
    .set_global_opts(
        title_opts=opts.TitleOpts(title="南昌各地房价对比折线图"),
        #提示线
        tooltip_opts=opts.TooltipOpts(trigger="axis"),
        #坐标轴配置
        xaxis_opts=opts.AxisOpts(
            is_show=True, #显示x轴
            type_="category", #显示离散数据类型
            axislabel_opts=opts.LabelOpts(rotate=45) #坐标轴标签旋转
        ),
        datazoom_opts=opts.DataZoomOpts(
            is_show=True,
            type_="slider",
            is_realtime=True,
            range_start=0,
            range_end=100
        )
    )
)
```

```python
[list(x) for x in zip(area_dfs_index_cn,list(building_cnts))]
```

```python
building_cnts=[int(value) for value in building_cnts]
pie=(
    Pie(
        init_opts=opts.InitOpts(
            width="800px",
            height="500px"
        )
    )
    .add("各地楼盘个数占比",[list(x) for x in zip(area_dfs_index_cn,building_cnts)])   
    .set_series_opts(
        label_opts=opts.LabelOpts(formatter="{b}:{c}")
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="南昌各地拥有楼盘占比",
            pos_top="30px"
        )
    )
)
```

```python
page.add(bar,line,pie)
page.render("render-building.html")
```
